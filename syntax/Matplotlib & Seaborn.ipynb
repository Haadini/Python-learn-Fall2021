```python
Mean = 10
Std = 2
Fig = Mean + Std * np.random.randn(500)

fig, ax=plt.subplots()

n, bins, patches = ax.hist(Fig, bins = 50, density = True)

Normal_Dist = ((1/(np.sqrt(2 * np.pi) * Std)) * np.exp(-0.5 * ((bins-Mean) / Std)**2))
ax.plot(bins, Normal_Dist)
plt.show()
```
